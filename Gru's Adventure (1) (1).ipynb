This game is based on the movie series <strong><i>Despicable Me</i></strong>. The main character <strong>Gru</strong> is a formerly evil supervillain turned spy. His most recent task is to thwart the terrible schemes of the three most deadly criminals in the world â€” <strong>Scarlet</strong> <strong>Overkill</strong>, <strong>Victor</strong> <strong>Perkins</strong>, and <strong>El</strong> <strong>Macho</strong>. Have a range of high-tech weaponry disposals, such as an inflation gun, a shrink ray, and a freeze ray. We must carefully employ these weapons to destroy each enemy and finish the quest.

You will take the role of <strong>Gru</strong>. Stopping <strong>Scarlet</strong> <strong>Overkill</strong> is the first step of your objective. She is renowned for her cutting-edge technologies and her mastery of fire. You must utilize your weapons carefully to fight her and stop her from destroying the city.

<strong>Victor</strong> <strong>Perkins</strong>, a master of disguise and cunning, awaits you in the second phase. He is a strong adversary, so be cautious not to fall for his traps. Use your weapons and your escape wisely to outwit him and foil his nefarious schemes. 

The third and final stage requires you to battle <strong>El</strong> <strong>Macho</strong>, a brutal tyrant with superhuman power. He is the boss of a formidable criminal gang and wants to rule the entire planet. To weaken him and stop his terror reign, use your weapons.

<u>Will you be able to vanquish the three evildoers and defend the world? You hold power to decide the future of the human race</u>.

```python

# "Gru's adventure against Scarlet Overkill, Victor Perkins, and El Macho"


# Define win and fail functions
def win():
    print("Congratulations! You have defeated all the enemies and won the game.")
    print("""                                        
#   #  ####  #    #    #    #  ####  #    # 
 # #  #    # #    #    #    # #    # ##   # 
  #   #    # #    #    #    # #    # # #  # 
  #   #    # #    #    # ## # #    # #  # # 
  #   #    # #    #    ##  ## #    # #   ## 
  #    ####   ####     #    #  ####  #    # 
  """)

def fail():
    print("You have been defeated. Better luck next time!")
    print("""
    
                                                                        
######  ###### ##### ##### ###### #####     #      #    #  ####  #    #    
#     # #        #     #   #      #    #    #      #    # #    # #   #     
######  #####    #     #   #####  #    #    #      #    # #      ####      
#     # #        #     #   #      #####     #      #    # #      #  #      
#     # #        #     #   #      #   #     #      #    # #    # #   #     
######  ######   #     #   ###### #    #    ######  ####   ####  #    #    
                                                                           
#    # ###### #    # #####    ##### # #    # ######                        
##   # #       #  #    #        #   # ##  ## #                             
# #  # #####    ##     #        #   # # ## # #####                         
#  # # #        ##     #        #   # #    # #                             
#   ## #       #  #    #        #   # #    # #                             
#    # ###### #    #   #        #   # #    # ######                        """)

# Main game loop
def stag1():
    # Define weapons list
    weapons = ["Freeze Ray", "Shrink Ray", "Inflation Gun"]
    
    # Define stage variables
    stage1 = "Fight against Scarlet Overkill"
    stage2 = "Escape from Victor Perkins' laboratory"
    stage3 = "Confront El Macho in his lair"

    # Define user's health and weapon variables
    current_weapon = ""
    health=100
    while health > 0:
        # Stage 1: Fight against Scarlet Overkill
            print("You are currently in stage 1: "+stage1)
            for i in weapons:
                print(i)
            
            current_weapon = input(prompt="select your weapons from the list, it is case sensitive")
            


            # Check if weapon is valid
            if current_weapon in weapons:
                print("You have selected the " + current_weapon + ".")

                # Nested conditional for weapon choice
                if current_weapon == "Freeze Ray":
                    print("You have frozen Scarlet Overkill and moved on to the next stage.")
                    stage1 = ""
                    stage2 = "Escape from Victor Perkins' laboratory"
                    stag2()
                    break
                elif current_weapon == "Shrink Ray":
                    print("You have shrunk Scarlet Overkill and moved on to the next stage.")
                    stage1 = ""
                    stage2 = "Escape from Victor Perkins' laboratory"
                    stag2()
                    break
                else:
                    print("You have used the Inflation Gun and defeated Scarlet Overkill.")
                    stage1 = ""
                    stage2 = "Escape from Victor Perkins' laboratory"
                    stag2()
                    break
                break
                    
            else:
                print("Invalid weapon choice. Please select a weapon from the list.")
                continue

# Stage 2: Escape from Victor Perkins' laboratory
def stag2():
        weapons = ["Freeze Ray", "Shrink Ray", "Inflation Gun"]

        # Define stage variables
        stage1 = "Fight against Scarlet Overkill"
        stage2 = "Escape from Victor Perkins' laboratory"
        stage3 = "Confront El Macho in his lair"
        print("You are currently in stage 2: " + stage2)
        print("You must escape the laboratory without being caught by Victor Perkins.")

        # Get user's choice of action
        action = input(prompt="select your options.\nHide or Fight")
        action=action.title()
        a=100
        while a==100:
        # Nested conditional for user's choice of action
            if action == "Hide":
                print("You have successfully hidden and escaped the laboratory.")
                stage2 = ""
                stage3 = "Confront El Macho in his lair"
                a +=1
                stag3()
                break

            elif action == "Fight":
                print("You have decided to fight Victor Perkins.")
                print("You have defeated Victor Perkins but have sustained injuries. Your health is now at " +".")
                stage2 = ""
                stage3 = "Confront El Macho in his lair"
                a +=1
                fail()
                break
                
            else:
                print("Invalid action. Please select 'hide' or 'fight'.")
                action = input(prompt="select your options.\nHide or Fight")
                action=action.title()
                continue

    # Stage 3: Confront El Macho in his lair
def stag3():
        weapons = ["Freeze Ray", "Shrink Ray", "Inflation Gun"]

        # Define stage variables
        stage1 = "Fight against Scarlet Overkill"
        stage2 = "Escape from Victor Perkins' laboratory"
        stage3 = "Confront El Macho in his lair"
        print("You are currently in stage 3: " + stage3)
        print("You must defeat El Macho to complete the game.")

       # Get user's choice of action
        action = input(prompt="select your option.\nConfront or Fight")
        action = action.title()

        # Nested conditional for user's choice of action
        a=100
        while a==100:
            if action == "Confront":
                print("You have successfully hidden and escaped the laboratory.")
                stage2 = ""
                stage3 = "Confront El Macho in his lair"
                a +=1
                win()
                break
                
            elif action == "Fight":
                print("You have decided to fight Victor Perkins.")
                fail()
                break
    
            else:
                print("Invalid action. Please select 'hide' or 'fight'.")
                action = input(prompt="select your option.\nConfront or Fight")
                action = action.title()
                continue

def strt():
    aa=1
    while aa>0:
        choice = input(prompt="Do you want to start.\nIf yes press 1 else 0")

        try:
            # converting choice for numerical comparison
            choice = int(choice)

        except:
            input(prompt = "I need a number in digits. Press enter to try again.\n")
            continue

        # conditional statement based on user input
        if choice == 1:
            stag1()
            break

        elif choice==0:
            fail()
            break

        # setting a different error message for inner and outer loops
        else:
            print("\nSomething went wrong.")
strt()
```
